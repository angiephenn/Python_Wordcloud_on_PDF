```python
import PyPDF2
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# C:\Users\angie\OneDrive\Desktop\Jupyter\raw files
```

```python
%matplotlib inline

# Open the PDF file
pdf_file = open(r'C:\Users\angie\OneDrive\Desktop\Jupyter\raw files\Angie_Phenn_CV.pdf', 'rb')

# Create a PDF reader object
pdf_reader = PyPDF2.PdfReader(pdf_file)

# Extract text from each page
text = ''
for page in pdf_reader.pages:
    text += page.extract_text()

# Generate word cloud with custom colors
wordcloud = WordCloud(colormap='BuPu').generate(text)  # Change the colormap as per your preference

# Note: https://matplotlib.org/stable/tutorials/colors/colormaps.html

# Configure the plot size and title
plt.figure(figsize=(10, 6))
plt.title('WordCloud')

# Display the generated word cloud using matplotlib
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()

# Close the PDF file
pdf_file.close()
```
